```python
# sep : 구분자
import pandas as pd
bank_data = pd.read_csv('Applied-Deep-Learning-with-Keras/Lesson01/data/bank.csv', sep =';')
```

```python
# 열 성분 중 y에 해당하는 열을 제거하겠다
feats = bank_data.drop('y', axis=1) 
# 열 성분 중 y에 해당하는 성분만 내포시킴
target = bank_data['y']
```

```python
# y가 사라진 것을 확인
feats
```

```python
# y만 출력됨을 확인
target
```

```python
# 행, 열이 몇개인지 확인 가능 = shape과 같음
print(f'Features table has {feats.shape[0]} rows and {feats.shape[1]} columns')
print(f'Target table has {target.shape[0]} rows')
```

```python
# data 파일 생성
!mkdir data
```

```python
# 활용했던 것을 다시 csv로 내보내고 싶을 경우
feats.to_csv('data/bank_data_feats.csv')
target.to_csv('data/bank_data_target.csv')
```
