```python
# pip istall pymysql ###다시 보기

import pymysql

db = pymysql.connect(
    # 접속할 컴퓨터의 주소
    host = '127.0.0.1',
    # 현재 구동중인 MySQL의 포트(서비스) 번호
    port = 3306,
    # 계정 이름
    user = 'bitai',
    # 비밀 번호
    passwd = '456123',
    # 사용하려는 스키마
    db = 'pydb'
)

# 후에 select 등을 했을 때 결과를 하나 하나 커서를 내려가며 순회할 수 있는 녀석
cursor = db.cursor()

# 실제 MySQL에서 사용하는 DB Query 문임
# create table 테이블 명( 원하는 필드와 원하는 타입을 작성)
# auto increment는 자동 증가
# unsigned란 부호가 없다는 의미 (즉 양수만 취급)
# varchar는 문자를 저장할 수 있는 배열
# primary key는 대표 키를 설정하는 키워드 (여기서는 id가 대표 키)
# 중복 되지 않음???

sql = """
    create table pydbtest(
        id int unsigned not null auto increment, 
        name varchar(20) not null,
        price int not null,
        primary key(id)
    )
"""

# 위의 SQL Query를 실행하는 코드
# 작업을 메모리 상에서 해놓고 메모리에 올려놓고 있음
cursor.execute(sql)

# DB에 실제 내용을 갱신하는 코드
# 실제 이부분은 I/O로 작업을 내리는 과정
db.commit()

# 왜 실행하자마자 적용되지 않고 commit()이란 것이 별도로 있나요?
# 파이썬에서 join()을 배웠었는데 같은 이유다.
# 컴퓨터 성능 높일 수 있는 좋은 방법 >>> 좋은 알고리즘을 쓴다.
# 다음으로 생각해야 하는 것은?
# I/O임 >>> I/O가 많이 발생하면 좋은 알고리즘이 아님
# 결론적으로 컴퓨터의 성능은 I/O와 직결된다.
# I/O를 최소화시켜야 하는데, DB는 데이터를 디스크에 저장하는 것.
# 레지스터 캐시 메모리 I/O 
# 디스크는 I/O >>> 작업할거 다하고 최종 갱신할 것만 commit해줘야 느려지지 않고 빠르게 가능.


# 작업을 완료했으므로 DB 연결을 끊는다.
# db를 열어둔 것이기 때문에 닫아줘야 함!
db.close()

print("Table Creation Success")
```

```python
import time
import math
import pymysql
import calendar
import datetime
import numpy as np
```

```python
sampleTime = np.arange(0, 10, 0.001) # 0.001c초마다 하나씩 생성

# 정규화 = 가우시안 분포 = 정규분포 (일반적인 분포)
noiseData = np.random.normal(size = len(sampleTime))

# f = 1/T
# T = 1/f
# 2 * pi * f * t = w * t
# 신호 만들기 -> 센서 데이터 수집한다 하면 이런식으로 만들면 됨.
voltage = 3 * np.cos(10 * sampleTime + (math.pi / 3)) + \
    2.7 * np.cos(7 * sampleTime) + \
    2.5 * np.cos(5 * noiseData * sampleTime)

current = 3 * np.cos(10 * sampleTime) + \
    2.7 * np.cos(7 * sampleTime) + \
    2.5 * np.cos(5 * noiseData * sampleTime)
```

```python
# 시간 정보 만들기
ts = calendar.timegm(time.gmtime())
ts = (sampleTime * 1000) + ts

sampleTime = np.array(sampleTime, dtype=np.float32)
voltage = np.array(voltage, dtype=np.float32)
current = np.array(current, dtype=np.float32)
```

```python
currTime = []

for i in range(len(sampleTime)):
    currTime.append(
        datetime.
            datetime.
                fromtimestamp(ts[i]).
                    strftime('%Y-%m-%d %H:%M:%S')
    )
```

```python
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = 'bitai',
    passwd = '456123',
    db = 'pydb'
)

cursor = db.cursor()


sql = """
    insert into sensor_data(
        voltage, current, vendor, name, reg_date
    ) values (%s, %s, %s, %s, %s)
"""
```

```python
try:
    with db.cursor() as cursor:
        for i in range(len(sampleTime)):
            cursor.execute(sql,
                (
                    str(voltage[i]),
                    str(current[i]),
                    'FMFire',
                    'F33N21BTN7',
                    currTime[i]
                )
            )
            
        db.commit()
finally:
    db.close()
    
print("Sucess to insert 10000 Sensor Data")

# 
```
